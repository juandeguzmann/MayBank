```python
import asyncio
import nest_asyncio
import sys
sys.path.insert(0, "..")
nest_asyncio.apply()

from backend.core.t212.client import TradingAppClient

async def fetch_sample():
    async with TradingAppClient() as client:
        return await client.get_orders(limit=5)

loop = asyncio.get_event_loop()
response = loop.run_until_complete(fetch_sample())
response
```
